# Exploratory Data Analysis (EDA)

Objective:
To explore customer demographics and behavior to identify patterns that may influence customer retention and churn.


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("../data/raw/Customer_Churn_Data_Large.xlsx")
```

```python
df['Gender'].value_counts()
df['Gender'].value_counts().plot(kind='bar')
plt.title("Gender Distribution")
plt.show()
```

- Gender distribution appears balanced.
- Gender alone may not strongly influence churn behavior.


```python
df['MaritalStatus'].value_counts().plot(kind='bar')
plt.title("Marital Status Distribution")
plt.show()
```

- Customers belong to different life stages.
- Life stage may impact financial needs and retention.


```python
df['IncomeLevel'].value_counts().plot(kind='bar')
plt.title("Income Level Distribution")
plt.show()
```

- Customers span multiple income groups.
- Income may influence product usage and retention risk.


## EDA Summary

EDA revealed meaningful demographic patterns that can be used to support customer retention strategies.
