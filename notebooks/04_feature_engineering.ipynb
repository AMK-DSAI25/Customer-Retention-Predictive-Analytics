# Feature Engineering

Note:
The dataset does not contain an explicit churn column.
Churn was defined using business rules based on customer engagement and account behavior.


```python
import pandas as pd

df = pd.read_csv("../data/processed/cleaned_customer_data.csv")
```

```python
df['Churn'] = 0
```

```python
df.loc[
    ((df['Age'] < 30) & (df['IncomeLevel'] == 'Low')) |
    ((df['MaritalStatus'] == 'Single') & (df['IncomeLevel'] == 'Low')),
    'Churn'
] = 1
```

```python
df['Churn'].value_counts()
```

```python
df.to_csv("../data/processed/featured_customer_data.csv", index=False)
```

Churn Definition:
As no behavioral or transactional data was available, churn was derived using demographic-based business assumptions.
Younger customers with lower income levels and single marital status were considered higher churn risk.
