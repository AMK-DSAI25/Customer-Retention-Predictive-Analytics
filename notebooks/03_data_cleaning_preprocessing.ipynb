# Data Cleaning and Preprocessing

Objective:
To clean and prepare customer data for churn analysis and modeling.


```python
import pandas as pd

df = pd.read_excel("../data/raw/Customer_Churn_Data_Large.xlsx")
```

```python
df.isnull().sum()
```

```python
df.fillna(method='ffill', inplace=True)
```

```python
df.to_csv("../data/processed/cleaned_customer_data.csv", index=False)
```

- Missing values were handled to ensure data consistency.
- Cleaned data was saved for further analysis.
